## Merge the Dataset


```python
merge=pd.merge(delivery,df,left_on='match_id', right_on='id')
merge.head()
```

```python
merge.describe()
```

## Top run getter Batsmans 

```python
run=delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head()
run
```

## Batsman with highest no. of sixes

```python
sixes = merge[merge['batsman_runs']==6]
c=sixes.groupby('batsman').count()['batsman_runs'].sort_values(ascending=False).head()
c
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=merge,y=c,x=c.index)
```

## Top wicket Takers Bowler

```python
D=delivery.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head()
D
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=merge,y=D,x=D.index)
plt.xticks(rotation=90)
plt.xlabel('Bowler')
plt.ylabel('Total wickets taken')
plt.show()
```
